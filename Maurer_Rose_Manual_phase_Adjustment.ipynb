```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
from IPython import display
from moviepy.editor import *

# %matplotlib notebook

# plt.figure(dpi=200)
# plt.figure(figsize=(10,5.625))
fig, ax = plt.subplots(figsize=(10,5.625))

k = 3
if k%2 == 1:
    angle = 180
elif k%2 == 0:
    angle = 360

# Set the range for the axis
ax.set_xlim(-2, 2)
ax.set_ylim(-1.125, 1.125)

# The x axis and y axis
# x axis
plt.plot([-2,2],[0,0],color = "w")
# y axis
plt.plot([0,0],[-1.125,1.125],color = "w")

a = 1

def transform(x,y,theta):
    X = (x*np.cos(theta))-(y*np.sin(theta))
    Y = (y*np.cos(theta))+(x*np.sin(theta))
    return X,Y

t = np.linspace(0,2*np.pi,360)
x_data = a*np.cos(k*t)*np.cos(t)
y_data = a*np.cos(k*t)*np.sin(t)
x_data, y_data = transform(x_data,y_data,30*np.pi/180)

files = []

for i in range(0,angle + 1,1):
    base = np.linspace(0,360,360)
    theta = base*i*(np.pi/180)
    r = np.sin(k*theta)
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    
    # For the graph sizing
    fig, ax = plt.subplots(figsize=(10,5.625))

    # Set the range for the axis
    ax.set_xlim(-2, 2)
    ax.set_ylim(-1.125, 1.125)

    # The x axis and y axis
    # x axis
    plt.plot([-2,2],[0,0],color = "w")
    # y axis
    plt.plot([0,0],[-1.125,1.125],color = "w")
    
    plt.plot(x,y,color = "#C71585",linewidth = 0.5)
    plt.plot(x_data,y_data,color = "#FFB6C1",linewidth = 2)
    plt.xticks([])
    plt.yticks([])
    plt.title("k = "+str(k))
    name = "k_is_"+str(k)+"Angle_is_"+str(i)+".jpg"
    plt.savefig(name,dpi = 250)
    files.append(name)
    plt.clf()
    
clip = ImageSequenceClip(files, fps = 10)
video_name = "video_"+str(k)+".mp4"
clip.write_videofile(video_name, fps = 25)
```
